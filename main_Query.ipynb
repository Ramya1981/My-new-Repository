```python
from Eda import AnemiaClassifier, HighBloodPressureClassifier, DeathEventClassifier

class HeartFailureAnalysis:
    def __init__(self, csv_file_path):
        self.csv_file_path = csv_file_path
        
    def analyze_data(self):
        # Create instances of classifiers
        anemia_classifier = AnemiaClassifier(self.csv_file_path)
        high_blood_pressure_classifier = HighBloodPressureClassifier(self.csv_file_path)
        death_event_classifier = DeathEventClassifier(self.csv_file_path)

        # Analyze Anemia
        accuracy_svm_anemia, precision_svm_anemia, recall_svm_anemia, f1_svm_anemia = anemia_classifier.analyze_anemia_svm()
        accuracy_knn_anemia, precision_knn_anemia, recall_knn_anemia, f1_knn_anemia=  anemia_classifier.analyze_anemia_knn()
        accuracy_rf_anemia, precision_rf_anemia, recall_rf_anemia, f1_rf_anemia=anemia_classifier.analyze_anemia_rf()
         # Display or store the results as needed
        print("Anemia - SVM Results:")
        print(f"Accuracy: {accuracy_svm_anemia}, Precision: {precision_svm_anemia}, Recall: {recall_svm_anemia}, F1 Score: {f1_svm_anemia}")
        
        print("\nAnemia - KNN Results:")
        print(f"Accuracy: {accuracy_knn_anemia}, Precision: {precision_knn_anemia}, Recall: {recall_knn_anemia}, F1 Score: {f1_knn_anemia}")
        
        print("\nAnemia - rf Results:")
        print(f"Accuracy: {accuracy_rf_anemia}, Precision: {precision_rf_anemia}, Recall: {recall_rf_anemia}, F1 Score: {f1_rf_anemia}")
        # Analyze High Blood Pressure
        accuracy_svm_high_bp, precision_svm_high_bp, recall_svm_high_bp, f1_svm_high_bp = high_blood_pressure_classifier.analyze_high_blood_pressure_svm()
        accuracy_knn_high_bp, precision_knn_high_bp, recall_knn_high_bp, f1_knn_high_bp = high_blood_pressure_classifier.analyze_high_blood_pressure_knn()
        accuracy_rf_high_bp, precision_rf_high_bp, recall_rf_high_bp, f1_rf_high_bp = high_blood_pressure_classifier.analyze_high_blood_pressure_rf()
        print("\nHigh Blood Pressure - SVM Results:")
        print(f"Accuracy: {accuracy_svm_high_bp}, Precision: {precision_svm_high_bp}, Recall: {recall_svm_high_bp}, F1 Score: {f1_svm_high_bp}")
        
        print("\nHigh Blood Pressure - knn Results:")
        print(f"Accuracy: {accuracy_knn_high_bp}, Precision: {precision_knn_high_bp}, Recall: {recall_knn_high_bp}, F1 Score: {f1_knn_high_bp}")
        
        print("\nHigh Blood Pressure - rf Results:")
        print(f"Accuracy: {accuracy_rf_high_bp}, Precision: {precision_rf_high_bp}, Recall: {recall_rf_high_bp}, F1 Score: {f1_rf_high_bp}")
        # Analyze Death Event
        accuracy_svm_death_event, precision_svm_death_event, recall_svm_death_event, f1_svm_death_event = death_event_classifier.analyze_DeathEvent_svm()
        accuracy_knn_death_event, precision_knn_death_event, recall_knn_death_event, f1_knn_death_event = death_event_classifier.analyze_DeathEvent_knn()
        accuracy_rf_death_event, precision_rf_death_event, recall_rf_death_event, f1_rf_death_event = death_event_classifier.analyze_DeathEvent_rf()

               
        
        
        print("\nDeath Event - SVM Results:")
        print(f"Accuracy: {accuracy_svm_death_event}, Precision: {precision_svm_death_event}, Recall: {recall_svm_death_event}, F1 Score: {f1_svm_death_event}")
        
        print("\nDeath Event - knn Results:")
        print(f"Accuracy: {accuracy_knn_death_event}, Precision: {precision_knn_death_event}, Recall: {recall_knn_death_event}, F1 Score: {f1_knn_death_event}")
        
        print("\nDeath Event - rf Results:")
        print(f"Accuracy: {accuracy_rf_death_event}, Precision: {precision_rf_death_event}, Recall: {recall_rf_death_event}, F1 Score: {f1_rf_death_event}")

if __name__ == "__main__":
    csv_file_path = "heart_failure_clinical_records_dataset-pcp.csv"
    
    # Create an instance of HeartFailureAnalysis
    heart_failure_analysis = HeartFailureAnalysis(csv_file_path)
    
    heart_failure_analysis.analyze_data()

```
